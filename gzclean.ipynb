```python
import pandas as pd
import re
```

```python
data_raw = pd.read_excel("./Data/广州.xlsx",header=0)
data_raw['flag'] = 0
r_date = r'\d{0,4}年{0,1}\d{1,2}月\d{1,2}日广州市新冠肺炎疫情情况'
r_olddate = r'截至\d{4}年\d{1,2}月\d{1,2}日24时广州市新冠肺炎疫情情况'
for index,row in data_raw.iterrows():
    flag = re.findall(r_date,str(row['标题']),re.I|re.M)
    flag_old = re.findall(r_olddate,str(row['标题']),re.I|re.M)
    if flag == [] and flag_old == []:
        flag = 0
    else:
        flag = 1
    print(flag)
    data_raw['flag'][index] = flag

data_new = data_raw[data_raw['flag']==1]
```

```python
#清楚无本土病例情况
r_trim = r'广州市新增境内确诊病例\d{1,}例'
r_oldtrim = r'广州市报告新冠肺炎新增确诊病例\d{1,}例'
r_oldnocase = r'广州市报告新冠肺炎新增确诊病例0例'

data_new['flag2'] = 0
for index, row in data_new.iterrows():
    content = str(row['正文'])
    flag = re.findall(r_trim,content,re.I|re.M)
    flag_old = re.findall(r_oldtrim,content,re.I|re.M)
    flag_oldnocase = re.findall(r_oldnocase,content,re.I|re.M)
    if flag == [] and flag_old == []:
        flag = 0
    
    elif flag_oldnocase != []:
        flag = 0
    else: 
        flag = 1
    data_new['flag2'][index]=flag

data_case = data_new[data_new['flag2']==1]
data_case = data_case.reset_index(drop=True)
print(data_case.index)
```

```python
# 清除附件代码
for i in range(len(data_case)):
    data_case['正文'][i] = data_case['正文'][i].split("附件")[0]
data_case.to_excel("./Data/gzcases.xlsx")
```

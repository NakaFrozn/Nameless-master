```python
import numpy as np
import pandas as pd
data = pd.read_csv("QAData_short.csv")
print(data.columns)
```

```python
data.columns=['id','province','city','age','gender','edu','earning','att','attcovid','attexp',
                '8','9','10','11','12','13','check','15','16','17','18','19',
                '20a','20b','20c','21a','21b','21c']
data_valid = data[data['check']!=1]
```

```python
value_mean = data_valid[['8','9','10','11','12','13','15','16','17','18','19']].mean()
print(value_mean)
value_attedu = data.groupby('edu')[['att','attcovid','attexp']].mean()
value_attedu.index=['初中及以下','高中/中专', '大专', '大学本科', '硕士及以上']
```

```python
from pyecharts.charts import Bar
index_list = list(value_attedu.index)
value_att = list(value_attedu['att'])
value_attcovid = list(value_attedu['attcovid'])
value_attexp = list(value_attedu['attexp'])
print(value_att)
bar = (
    Bar()
    .add_xaxis(index_list)
    .add_yaxis("对个人隐私保护的关注程度",value_att)
    .add_yaxis("对疫情期间个人隐私保护的关注程度",value_attcovid)
    .add_yaxis("对流调信息的关注程度",value_attexp)
    .set_global_opts(title_opts={"text":"主标题"})
)
bar.render_notebook()
```
